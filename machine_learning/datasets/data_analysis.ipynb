# Data Analysis untuk Project SafeFood

Import Library yang dibutuhkan

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Analisis Data Donor

```python
# Membaca data donor
donor_data = pd.read_csv('cleaned_data/data_donor_cleaned.csv')
print("Data Donor berhasil dibaca.")
```

#### 1. Distribusi Jenis Makanan

```python
# Menampilkan proporsi jenis makanan yang didonasikan
donor_data['jenis_makanan_flat'] = donor_data['jenis_makanan'].apply(lambda x: x[0] if isinstance(x, list) else x)

# Menghitung proporsi jenis makanan
jenis_makanan_counts = donor_data['jenis_makanan_flat'].value_counts()

# Plot Pie Chart
plt.figure(figsize=(8, 6))
jenis_makanan_counts.plot.pie(autopct='%1.1f%%', startangle=90, cmap='Set3')
plt.title("Distribusi Jenis Makanan yang Didonasikan")
plt.ylabel('')
plt.show()

# Plot Bar Chart
plt.figure(figsize=(8, 6))
sns.barplot(x=jenis_makanan_counts.index, y=jenis_makanan_counts.values, palette="Set3")
plt.title("Distribusi Jenis Makanan yang Didonasikan")
plt.xlabel("Jenis Makanan")
plt.ylabel("Jumlah Donasi")
plt.show()
```
